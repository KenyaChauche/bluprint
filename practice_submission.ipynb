```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

```python
train = pd.read_csv("./datasets/train.csv")
test = pd.read_csv("./datasets/test.csv")
```

```python
train.head()
```

```python
train.columns
```

```python
X = train[["1st Flr SF", "2nd Flr SF", "Lot Area", "Full Bath", "Half Bath", "Bedroom AbvGr"]]

y = train["SalePrice"]
```

```python
model = LinearRegression()

model.fit(X, y)
```

```python
model.coef_
```

```python
X_test = test[["1st Flr SF", "2nd Flr SF", "Lot Area", "Full Bath", "Half Bath", "Bedroom AbvGr"]]
```

```python
preds = model.predict(X_test)
```

```python
test["SalePrice"] = preds 
```

```python
test.head()
```

```python
submission = test[["Id", "SalePrice"]]
```
