### Imports and Data Reading

```python
import pandas as pd
import missingno as msno
import numpy as np 
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv("./datasets/train.csv")
```

```python
pd.set_option("display.max_columns", 100)
pd.set_option("display.max_rows", 100)
train.head().T
```

```python
msno.bar(train); 
```

```python
# make all column names lower case and with underscores 
col_names = list(train.columns)

col_names = [col_names[x].lower() for x in range(len(col_names))]

col_names = [col_names[x].replace(" ", "_") for x in range(len(col_names))]

train.columns = col_names
```
